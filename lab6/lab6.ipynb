Import Pandas, Numpy, and create the associated data frame

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sales_csv=pd.read_csv("company_sales_data.csv")
sales_dataFrame= pd.DataFrame(sales_csv)
print(sales_dataFrame.info())
```

Exercise 1: Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties: –

X label name = Month Number
Y label name = Total profit


```python
#sales_dataFrame.plot(kind='line', x='month_number', y='total_profit', title='Company Profit Per Month')
#plt.xlabel('Month Number')
#plt.ylabel('Total Profit')
#plt.show()

# Set Seaborn style
#sns.set_style("whitegrid")

# Line Plot
sns.lineplot(data=sales_dataFrame, x='month_number', y='total_profit')
plt.title('Company Profits Per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(sales_dataFrame['month_number'])
plt.show()
```

Exercise 2: Get total profit of all months and show line
plot with the following Style properties
Generated line plot must include following Style properties: –
● Line Style dotted and Line-color should be red
● Show legend at the lower right location.
● X label name = Month Number
● Y label name = Sold units number
● Add a circle marker.
● Line marker color as read
● Line width should be 3

```python

plt.plot(sales_dataFrame['month_number'], sales_dataFrame['total_units'], marker='o', color='red', linestyle='dashed', linewidth=3, label='Sold Units Data of Last Year', markerfacecolor='black' )
plt.title('Company Sales Data of Last Year')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='lower right')
plt.xticks(sales_dataFrame['month_number'])
plt.show()
```

Exercise 3: Read all product sales data and show it
using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate
Plotline for each product ).

```python
import random

for col in sales_dataFrame.columns[1:6]:
    plt.plot(sales_dataFrame['month_number'], sales_dataFrame[col], marker='o', color = "#{:06x}".format(random.randint(0, 0xFFFFFF)), linewidth=3, label=col)

plt.title('Sales Data')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='upper left')

plt.xticks(sales_dataFrame['month_number'])
plt.show()
```

Exercise 4: Read toothpaste sales data of each month
and show it using a scatter plot
Also, add a grid in the plot. gridline style should “–“

```python
plt.scatter(sales_dataFrame['month_number'], sales_dataFrame['toothpaste'],   label='Toothpaste Sales')
plt.title('Toothpaste Sales Over Last Year')
plt.xlabel('Month Number')
plt.ylabel('Sold Units Number')
plt.legend(loc='upper left')
plt.grid(True, linestyle='--')
plt.xticks(sales_dataFrame['month_number'])

plt.show()
```

Exercise 7: Read the total profit of each month and
show it using the histogram to see the most common
profit ranges

```python
sns.histplot(data=sales_dataFrame, x='total_profit', bins=np.linspace(sales_dataFrame['total_profit'].min(), sales_dataFrame['total_profit'].max(), 12 ))
plt.xticks(np.linspace(sales_dataFrame['total_profit'].min(), sales_dataFrame['total_profit'].max(), 12 ))
plt.xticks(rotation=90)
plt.show()
```

Exercise 8: Calculate total sale data for last year for
each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

```python
sales_sums = sales_dataFrame.iloc[:, 1:6].sum()



plt.pie(sales_sums, labels=sales_dataFrame.columns[1:6])

plt.title('Sales')

plt.legend(loc='lower right')
plt.show()
```
