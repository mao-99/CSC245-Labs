1. Write a Pandas program to print a concise summary of the dataset (titanic.csv).
(Also imported pandas)

```python
import pandas as pd

titanic_csv=pd.read_csv("titanic.csv")
titanic_dataFrame= pd.DataFrame(titanic_csv)
print(titanic_dataFrame.info())
```

2.Write a Pandas program to extract the column labels, shape and data types of the dataset (titanic.csv).

```python
print("Columns:", titanic_dataFrame.columns.to_list())
print("Shape:", titanic_dataFrame.shape)
print("Data Types:\n", titanic_dataFrame.dtypes)
```

3. Write a Pandas program to create a Pivot table with multiple indexes from the data set of titanic.csv.

```python
##The default aggregate function of pivot table is Average
titanic_pivot_table = titanic_dataFrame.pivot_table(index=["class", "sex" ], values="age", );
print(titanic_pivot_table)
```

4. Write a Pandas program to create a Pivot table and find survival rate by gender on various classes.

```python
titanic_pivot_table = titanic_dataFrame.pivot_table(index=[ "class", 'sex' ],   values="survived");
print(titanic_pivot_table)
```

5. Write a Pandas program to create a Pivot table and find survival rate by gender.

```python
titanic_pivot_table = titanic_dataFrame.pivot_table(index='sex',   values="survived");
print(titanic_pivot_table)
```

6. Write a Pandas program to create a Pivot table and find survival rate by gender, age wise of various classes.

```python

titanic_pivot_table = titanic_dataFrame.pivot_table(index=[ 'sex', 'class' ], columns='age',  values="survived");
print(titanic_pivot_table)
```

7. Write a Pandas program to partition each of the passengers into four categories based on their age.

```python
titanic_dataFrame["group_by_age"]=pd.cut(titanic_dataFrame["age"], bins=[0, 10, 30, 60, 80])

print(titanic_dataFrame)

```

8. Write a Pandas program to create a Pivot table and count survival by gender, categories wise age of various classes.

```python
titanic_pivot_table = titanic_dataFrame.pivot_table(index=[ 'sex', 'class' ], columns='group_by_age',  values="survived", aggfunc="count");
print(titanic_pivot_table)
```

9. Write a Pandas program to create a Pivot table and find survival rate by gender, age of the different categories of various classes.

```python
titanic_pivot_table = titanic_dataFrame.pivot_table(index=[ 'sex', 'group_by_age' ], columns='class',  values="survived", aggfunc="mean");
print(titanic_pivot_table)
```
